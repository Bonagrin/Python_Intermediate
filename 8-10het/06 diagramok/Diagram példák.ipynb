```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.plot(np.array([3, 7, 6, 12, 16]))
plt.show()
```

```python
plt.plot(np.array([1, 3, 4, 8, 9]), np.array([3, 7, 6, 12, 16]))
plt.show()
```

```python
plt.plot(
    np.array([1, 3, 4, 8, 9]),
    np.array([3, 7, 6, 12, 16]),
    marker='o',
    markersize=20,
    markeredgecolor='black',
    markerfacecolor='red',
    linestyle='dashed',
    linewidth=5,
    color='green'
)
plt.show()
```

```python
plt.scatter(np.array([1, 1, 4, 4, 9]), np.array([3, 7, 7, 12, 16]))
plt.show()
```

```python
np.random.seed(197777)
plt.scatter(
    x = np.random.randint(100, size=(100)),      # x-coordinate
    y = np.random.randint(100, size=(100)),      # y-coordinate
    c = np.random.randint(100, size=(100)),      # color
    cmap = 'nipy_spectral',                      # color map
    s = 10 * np.random.randint(100, size=(100)), # size
    alpha = 0.5,                                 # transparency
)
plt.colorbar()
plt.show()
```

```python
plt.bar(['alma', 'narancs', 'bann', 'meggy', 'citrom'], np.array([1, 8, 4, 9, 3]))
plt.show()
```

```python
plt.barh(['alma', 'narancs', 'bann', 'meggy', 'citrom'], np.array([1, 8, 4, 9, 3]))
plt.show()
```

```python
plt.pie(np.array([1, 8, 4, 9, 3]), labels=['alma', 'narancs', 'bann', 'meggy', 'citrom'])
plt.show()
```

```python
plt.pie(
    np.array([1, 8, 4, 9, 3]),
    labels=['alma', 'narancs', 'bann', 'meggy', 'citrom'],
    explode=[0.0, 0.0, 0.2, 0.0, 0.0]
)
plt.show()
```

```python
np.random.seed(197777)
plt.hist(np.random.normal(100, 15, 1000), bins=50)
plt.show()
```

```python
plt.plot(np.array([1, 8, 4, 9, 3]))
plt.plot(np.array([7, 6, 2, 7, 8]))
plt.show()
```

```python
gymlcsk = ['alma', 'narancs', 'bann', 'meggy', 'citrom']
szmok = np.array([1, 8, 4, 9, 3])

plt.subplot(2, 2, 1)
plt.plot(szmok)

plt.subplot(2, 2, 2)
plt.scatter(gymlcsk, szmok)

plt.subplot(2, 2, 3)
plt.barh(gymlcsk, szmok)

plt.subplot(2, 2, 4)
plt.pie(szmok, labels=gymlcsk)

plt.show()
```

```python
plt.plot(np.array([1, 3, 4, 8, 9]), np.array([3, 7, 6, 12, 16]))
plt.title('Vonaldiagram', fontdict={'family':'serif','color':'blue','size':20})
plt.xlabel('x koordinta')
plt.ylabel('y koordinta')
plt.grid()
plt.show()
```

```python
gymlcsk_df = pd.DataFrame({
    'nv': ['alma', 'narancs', 'bann', 'meggy', 'citrom'],
    'rtk': [1, 8, 4, 9, 3],
})
gymlcsk_df.plot(kind='barh', x='nv', y='rtk')
plt.show()
```

```python
penguins = sns.load_dataset("penguins")
g = sns.pairplot(penguins, hue="species")
g.map_lower(sns.kdeplot)
plt.show()
```
