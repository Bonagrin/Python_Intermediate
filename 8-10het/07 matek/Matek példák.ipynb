```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100)
y = x * np.sin(x) + 4

ax = plt.figure().add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')

plt.plot(x, y)
plt.show()
```

```python
import scipy.optimize
import math

scipy.optimize.root(lambda x: x * math.sin(x) + 4, 0).x
```

```python
x = np.linspace(-5, 5, 100)
y = x**2 - 6*x + 2

ax = plt.figure().add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')

plt.plot(x, y)
plt.show()
```

```python
minimum = scipy.optimize.minimize(lambda x: x**2 - 6*x + 2, 1)
print(minimum.x)
print(minimum.fun)
```
