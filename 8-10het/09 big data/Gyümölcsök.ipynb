```
gyümölcsök_df = spark.createDataFrame([
  ('alma', 3),
  ('banán', 2),
  ('narancs', 5),
  ('alma', 4),
  ('banán', -1),
  ('alma', 1)
]).toDF('nev', 'darab')
```

```
gyümölcsök_df.show()
```

```
gyümölcsök_df.createOrReplaceTempView('gyumolcsok')
spark.sql('SELECT nev, sum(darab) AS darab FROM gyumolcsok WHERE darab > 0 GROUP BY nev ORDER BY darab DESC').show()
```

```
from pyspark.sql.functions import sum, col, desc
 
gyümölcsök_df \
  .select('nev', 'darab') \
  .where(col('darab') > 0) \
  .groupBy('nev') \
  .agg(sum('darab').alias('darab')) \
  .sort(desc('darab'))\
  .show()
```

```
import pyspark.pandas as ps

gyumolcsok_pandas = ps.DataFrame({
  'nev': ['alma', 'banán', 'narancs', 'alma', 'banán', 'alma'],
  'darab': [3, 2, 5, 4, -1, 1],
})
gyumolcsok_pandas_filtered = gyumolcsok_pandas[gyumolcsok_pandas['darab'] > 0]
gyumolcsok_pandas_grouped = gyumolcsok_pandas_filtered.groupby('nev').sum().reset_index()
gyumolcsok_pandas_sorted = gyumolcsok_pandas_grouped.sort_values('darab', ascending=False)
print(gyumolcsok_pandas_sorted)
```
